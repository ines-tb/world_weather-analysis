```python
# Import dependencies
import pandas as pd
import numpy as np
from citipy import citipy
# Import OpenWeatherMap API key
from config import weatherApiKey
import requests
import os
```

```python
# Create a set of random latitude and longitude combinations
# ---------------------------------------------------------------

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
# Zip the latitudes and longitudes as tuples
latLngs = zip(lats, lngs)
# Covert the latitudes and longitudes from zip object into a list
coordinates = list(latLngs)
```

```python
# Get nearest city using citipy
# ---------------------------------------------

# Create a list for holding the found cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count
len(cities)
```

```python
# Retrieve weather information from OpenWeatherMap API
# ---------------------------------------------------------------

# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weatherApiKey

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create an empty list to hold the weather data.   
cityData = []
# Create counters.
recordCount = 1
setCount = 1
# Loop through all the cities in our list.
for i, city in enumerate(cities):
    
    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        setCount += 1
        recordCount = 1
    # Create endpoint URL with each city
    cityUrl = url + "&q=" + city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {recordCount} of Set {setCount} | {city}")
    
    # Add 1 to the record count.
    recordCount += 1

    # Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        cityWeather = requests.get(cityUrl).json()
        # Parse out the needed data
        cityCountry = cityWeather["sys"]["country"]
        cityLat = cityWeather["coord"]["lat"]
        cityLng = cityWeather["coord"]["lon"]
        cityMaxTemp = cityWeather["main"]["temp_max"]
        cityHumidity = cityWeather["main"]["humidity"]
        cityClouds = cityWeather["clouds"]["all"]
        cityWind = cityWeather["wind"]["speed"]
        cityDescr = cityWeather["weather"][0]["description"]
        
        # Try to get rain in the last hour, otherwise set the value to 0
        try:
            cityRain = cityWeather["rain"]["rain.3h"]
        except KeyError:
            cityRain = float(0)
            pass

        # Try to get snow in the last hour, otherwise set the value to 0
        try:
            citySnow = cityWeather["snow"]["snow.3h"]
        except KeyError:
            citySnow = float(0)
            pass

        # Append the city information into city_data list.
        cityData.append({
            "City" : city.title(),
            "Country" : cityCountry,
            "Lat" : cityLat,
            "Lng" : cityLng,
            "Max Temp" : cityMaxTemp,
            "Humidity" : cityHumidity,
            "Cloudiness" : cityClouds,
            "Wind Speed": cityWind,
            "Current Description": cityDescr,
            "Rain (inches)": cityRain,
            "Snow (inches)": citySnow})

    # If an error different from 'rain or snow not found' is experienced, skip the city.
    except:
        print(f"City '{city}' not found. Skipping...")
        pass


# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Add data to a DataFrame
# --------------------------

# Convert the array of dictionaries to a Pandas DataFrame.
cityDataDF = pd.DataFrame(cityData)
# Set correct order
columnOrder = ["City", "Country", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description", "Rain (inches)", "Snow (inches)"]
cityDataDF = cityDataDF[columnOrder]

# Display city DataFrame
cityDataDF.head(10)
```

```python
# Challenge question: How many cities have recorded rainfall or snow?
print("Raining cities:")
print(cityDataDF.loc[(cityDataDF["Rain (inches)"] != 0)]["City"].count())
print("\nSnowing cities:")
print(cityDataDF.loc[(cityDataDF["Snow (inches)"] != 0)]["City"].count())
```

```python
# Save DataFrame to a csv file
# --------------------------------------

# Create the output file (CSV).
outputDataFile = os.path.join("data", "WeatherPy_database.csv")
# Export the City_Data into a CSV.
cityDataDF.to_csv(outputDataFile, index_label="City_ID")
```
